```python
import pandas as pd
data = pd.read_csv("cgpaddfdf.csv")
X = data['CGPA']
Y = data['IQ']
data.shape
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,Y)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
data_scaled
```

```python

wcss = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, random_state=42)
    km.fit(data_scaled)
    wcss.append(km.inertia_)


wcss
```

```python
plt.plot(range(1,11), wcss)
```

```python


km = KMeans(n_clusters=3, random_state=42)
y_means = km.fit_predict(data_scaled)
y_means

plt.scatter(data_scaled[y_means == 0,0],data_scaled[y_means == 0,1], color='blue')
plt.scatter(data_scaled[y_means == 1,0],data_scaled[y_means == 1,1], color='red')
plt.scatter(data_scaled[y_means == 2,0],data_scaled[y_means == 2,1], color='green')
```
