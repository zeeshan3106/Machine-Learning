```python
import pandas as pd
data = pd.read_csv("melbourne_wedding_spending.csv")
# data = data.drop(['CafeID','Monthly_Coffee_Beans_kg','Google_Rating'],axis = 1)
data
X = data.iloc[:,0:4].values
X
```

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
plt.figure(figsize=(10, 7))
dendrogram = sch.dendrogram(sch.linkage(data, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
ag = AgglomerativeClustering(
    n_clusters = 3,
    linkage = 'ward',
)
training = ag.fit_predict(X)
X[training==0].shape


```

```python
plt.scatter(X[training==1,0],X[training==1,1])
plt.scatter(X[training==2,0],X[training==2,1])
plt.scatter(X[training==0,0],X[training==0,1])
```
