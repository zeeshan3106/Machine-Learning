```python
import pandas as pd
data = pd.read_csv("customer_clustering_dataset.csv")
data = data.drop(['CustomerID'],axis = 1)
data
```

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
plt.figure(figsize=(10, 7))
dendrogram = sch.dendrogram(sch.linkage(data, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
ag = AgglomerativeClustering(
    n_clusters = 3,
    linkage = 'single',
)
training = ag.fit_predict(data)
data[training==1].shape
```

```python
plt.scatter(data[training==1],data[training==1])
plt.scatter(data[training==2],data[training==2])
plt.scatter(data[training==0],data[training==0])
```
