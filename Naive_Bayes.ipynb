```python
import pandas as pd
data = pd.read_csv("tennis_playing_records.csv")
data.head(1)
```

**`Applying Naive Bayes `**

```python
out = data['Outlook'].value_counts()
temp = data['Temperature'].value_counts()
hum = data['Humidity'].value_counts()
wind = data['Wind'].value_counts()
print(out,"\n",temp,"\n",hum,"\n",wind)
```

```python
outlook = pd.crosstab(data['Outlook'],data['PlayTennis'])
outlook
```

```python
overcno= outlook .iloc[0,0]/outlook .iloc[:,0].sum()

rainno = outlook .iloc[1,0]/outlook .iloc[:,0].sum()

Sunnyno = outlook .iloc[2,0]/outlook .iloc[:,0].sum()


overcyes = outlook .iloc[0,1]/outlook .iloc[:,1].sum()

rainyes = outlook .iloc[1,1]/outlook .iloc[:,1].sum()
rainyes
Sunnyyes = outlook .iloc[2,1]/outlook .iloc[:,1].sum()
```

```python
temp = pd.crosstab(data['Temperature'],data['PlayTennis'])
temp
```

```python
tmpcolno = temp.iloc[0,0]/temp.iloc[:,0].sum()
tmpcolno
tmphotno = temp.iloc[1,0]/temp.iloc[:,0].sum()
tmphotno
tmpmildno = temp.iloc[2,0]/temp.iloc[:,0].sum()
tmpmildno

tmpcolyes = temp.iloc[0,1]/temp.iloc[:,1].sum()
tmpcolyes
tmphotyes = temp.iloc[1,1]/temp.iloc[:,1].sum()
tmphotyes
tmpmildyes = temp.iloc[2,1]/temp.iloc[:,1].sum()
tmpmildyes
```

```python
hum = pd.crosstab(data['Humidity'],data['PlayTennis'])
hum
```

```python
humhighno = hum.iloc[0,0]/hum.iloc[:,0].sum()
humhighno
humhighyes = hum.iloc[1,0]/hum.iloc[:,1].sum()
humhighyes

humnormalo = hum.iloc[0,0]/hum.iloc[:,0].sum()
humnormalo
humnormalyes = hum.iloc[1,0]/hum.iloc[:,1].sum()
humnormalyes
```

```python
wind = pd.crosstab(data['Wind'],data['PlayTennis'])
wind
```

```python
WindStrongno = wind.iloc[0,0]/wind.iloc[:,0].sum()
WindStrongno
WindStrongYes = wind.iloc[1,0]/wind.iloc[:,1].sum()
WindStrongYes
Windweekno = wind.iloc[0,0]/wind.iloc[:,0].sum()
WindStrongno
WindweekYes = wind.iloc[1,0]/wind.iloc[:,1].sum()
WindStrongYes
```

```python
p = data['PlayTennis'].value_counts()
py = p['Yes']
py
pn= p['No']
```

```python
pyes = py*Sunnyyes*tmphotyes*humhighyes*WindStrongYes
pyes
```

```python
pno = pn*Sunnyno*tmphotno*humhighno*WindStrongno
pno
```

Normalizing


```python
pyespercentage = pyes/(pyes+pno)
pyespercentage = pyespercentage*100
pnopercentage = pno/(pyes+pno)
pnopercentage = pnopercentage*100
if (pyespercentage>pnopercentage):
  print(pyespercentage,"%","\n","yes there would definiately a wonderful Match")
else:
  print(pnopercentage,"%","\n","No Match In these Conditions.")
```
