```python
X
error = 0.125
alpha = 0.5*np.log((1-error)/(error))
alpha
```

```python
new_weight = 0.125 * np.exp(alpha)
new_weight
C_new_weight = 0.125*np.exp(-alpha)
C_new_weight
```

```python
import numpy as np

def function(row, apple, ball):
    # Check the specific row's Y and Y_pred values
    if row['Y'] == row['Y_pred']:
        return ball                       # Correct prediction (C_new_weight)
    else:
        return 0.125 * np.exp(apple)      # Incorrect prediction (new_weight)

apple = alpha
ball = C_new_weight

# Pandas loops through row-by-row automatically
X['new_weight'] = X.apply(function, axis=1, args=(apple, ball))
X
```

```python
X['new_weight'] = X['new_weight']/X['new_weight'].sum()
X.head(1)
```

```python
X['Upper']=np.cumsum(X['new_weight'])
X['Lower']=X['Upper']-X['new_weight']

X
```

```python
def function(X):
  incides = []
  for i in range(X.shape[0]):
    a = np.random.random()
    for index, row in X.iterrows():

      if row['Upper']>a and a > row['Lower']:
        incides.append(index)
  return incides

X = X
iv = function(X)
iv


```
