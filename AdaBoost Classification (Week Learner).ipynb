```python
import pandas as pd

data = {
    'Rooms': [3, 2, 4, 1, 3, 5, 2, 4],
    'Distance_to_CBD': [4.2, 12.5, 5.1, 2.8, 18.0, 8.3, 22.1, 3.5],
    'Is_Expensive': [1, 0, 1, 0, 0, 1, 0, 1]
}
data = pd.DataFrame(data)
data
X = data.drop('Is_Expensive',axis=1)
y = data['Is_Expensive']
y
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(max_depth = 1)
X_train = tree.fit(X,y)
y_pred = tree.predict(X)
from sklearn.metrics import accuracy_score
accuracy_score(y,y_pred)*100

```
